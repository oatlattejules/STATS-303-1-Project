```python
data = pd.read_csv('heart_2022_no_nans.csv')
data.columns
```

```python
data_selected = data[['State','GeneralHealth', 'PhysicalActivities', 'SleepHours', 'HadHeartAttack', 'HadAngina', 'HadStroke',
                      'HadAsthma','HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
                      'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'SmokerStatus', 'ECigaretteUsage',
                      'AgeCategory', 'WeightInKilograms', 'BMI', 'AlcoholDrinkers']]

data_selected.nunique()
```

### 4. What regions of the US are more/less at risk for heart disease?

```python
final_data = data_selected[['State', 'HadHeartAttack', 'HadAngina']]
heartattack_counts = final_data['HadHeartAttack'].value_counts()
angina_counts = final_data['HadAngina'].value_counts()

print(heartattack_counts)
print(angina_counts)
```

```python
count_heartattack = final_data.groupby(['State', 'HadHeartAttack']).size().reset_index(name='Count')
pivot_heartattack = count_heartattack.pivot(index='State', columns='HadHeartAttack', values='Count').fillna(0)
pivot_heartattack
```
